```python
#Inicializando novo dataframe, dessa vez com informações de resultados de estudantes.
estudantes = pd.read_csv("https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv")
estudantes
```

```python
#Exibe informações das notas
estudantes["writing score"].describe()
```

```python
#Exibe informações da educação dos pais
estudantes["parental level of education"].describe()
```

```python
#Exibe os valores possíveis daquela coluna
estudantes["parental level of education"].unique()
```

```python
#Conta o total de vezes que o valor aparece na lista
estudantes["parental level of education"].value_counts()
```

```python
#Calcula média e armazena numa variável
media_escrita = estudantes["writing score"].mean()
#Calcula quão distante da média está o valor (acima é positivo, abaixo é negativo)
estudantes["writing score"].map(lambda p: p - media_escrita)
#Retorna uma série onde todos os valores foram transformados de acordo com a função
```

```python
#Faz a mesma coisa do anterior, mas usando a função apply() e para todas as médias
def nova_media(linha):
    linha["writing score"] = linha["writing score"] - estudantes["writing score"].mean()
    linha["math score"] = linha["math score"] - estudantes["math score"].mean()
    linha["reading score"] = linha["reading score"] - estudantes["reading score"].mean()
    return linha

estudantes.apply(nova_media, axis="columns")
```

```python
#Usando groupby pra simular o value_counts
estudantes.groupby("gender").gender.count()
```

```python
#Pegando a menor nota de matemática de cada nível de educação dos pais
estudantes.groupby("parental level of education")["math score"].min()
```

```python
#Pegando a maior nota de matemática de cada nível de educação dos pais
estudantes.groupby("parental level of education")["math score"].max()
```

```python
#Agrupa os estudantes por nível de educação dos pais e raça, retornando as notas de quem teve a média mais alta
def maior_media(df):
    return df.loc[(
        (df["math score"] + df["reading score"] + df["writing score"])/3
    ).idxmax()]

estudantes.groupby(["parental level of education", "race/ethnicity"]).apply(maior_media)
```

```python
#Agrupa os estudantes por nível de educação dos pais e raça, retornando as notas de quem teve a menor média
def menor_media(df):
    return df.loc[(
        (df["math score"] + df["reading score"] + df["writing score"])/3
    ).idxmin()]

estudantes.groupby(["parental level of education", "race/ethnicity"]).apply(menor_media)
```

```python
#Agrupa os estudantes por gênero e nível de educação dos pais e retorna a nota mais alta e mais baixa em cada grupo, 
#junto da quantidade de pessoas naquele grupo
estudantes.groupby(["parental level of education", "gender"])[["math score", "reading score", "writing score"]].agg([len, min, max])
```

```python
#Filtra os estudantes por educação dos pais e se houve preparatório ou não e conta quantos estudantes estão em cada grupo
resumo = estudantes.groupby(["parental level of education", "test preparation course", "gender"])["gender"].agg([len])

#Agora organiza os valores em ordem crescente 
resumo.sort_values(by="len")
```

```python
#Mesma coisa, mas em ordem decrescente
resumo.sort_values(by="len", ascending=False)
```

```python
#Organizando agora pelo sexo e pela quantidade de pessoas
resumo.sort_values(by=["gender", "len"])
```

```python
#Renomeando colunas
estudantes = estudantes.rename(columns={
    "race/ethnicity": "race_ethnicity",
    "parental level of education": "parental_education",
    "test preparation course": "preparation_course",
    "math score": "math_score",
    "reading score": "reading_score",
    "writing score": "writing_score"
})
#Para renomear o índice é só mudar o columns para index
estudantes
```
